```python
import os
import csv
import glob
import numpy as np
import pandas as pd
import matplotlib as plt
from functools import reduce

```

```python
csv_files = glob.glob('*.csv')
print(csv_files)
```

```python
list_data = []

for filename in csv_files:
    data = pd.read_csv(filename,index_col=None, header=0)
    list_data.append(data)
    
list_data
```

```python
df_final = reduce(lambda left,right: pd.merge(left,right,on="CODE"), list_data)
df_final
```

```python
#defined a function to drop suffixes

def drop_y(df_final):
    # list comprehension of the cols that end with '_y'
    to_drop = [x for x in df_final if x.endswith('_y')| x.endswith('_x')]
    df_final.drop(to_drop, axis=1, inplace=True)

#merge list_data into a dataframe
df_final = reduce(lambda left,right: pd.merge(left,right,on="CODE"), list_data)

drop_y(df_final)

df_final
```

```python
#Drop duplicate colums

Cols = list(df_final.columns)

for i,item in enumerate(df_final.columns):
    if item in df_final.columns[:i]: Cols[i] = "NAICS DESCRIPTION"
df_final.columns = Cols
df_dropdup = df_final.drop("NAICS DESCRIPTION",1)
df_dropdup
```

```python
#list of column names
cols = list(df_dropdup.columns.values)
cols
```
